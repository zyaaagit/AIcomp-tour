```python
def get_hotel_item(hotel_list):
    hotel_url = hotel_list['url']
    proxy = random.choice(proxy_list)
    proxies = {
        'http': 'http://' + proxy,
        'https': 'http://' + proxy,
    }
    r = requests.get(hotel_url,headers= ua(),proxies=proxies)
    html = etree.HTML(r.text)
    hotel_open_time = html.xpath('//span[@class="label"][contains(text(), "入住时间")]/following-sibling::span/strong/text()')
    hotel_open_time = hotel_open_time[0].strip() if hotel_open_time else ""

    hotel_out_time = html.xpath('//span[@class="label"][contains(text(), "离店时间")]/following-sibling::span/strong/text()')
    hotel_out_time = hotel_out_time[0].strip() if hotel_out_time else ""

    hotel_constructed_year = html.xpath('//span[@class="label"][contains(text(), "建成于")]/following-sibling::span/strong/text()')
    hotel_constructed_year = hotel_constructed_year[0].strip() if hotel_constructed_year else ""
    
    hotel_renovated_year = html.xpath('//span[@class="label"][contains(text(), "翻修于")]/following-sibling::span/strong/text()')
    hotel_renovated_year = hotel_renovated_year[0].strip() if hotel_renovated_year else ""

    hotel_size = html.xpath('//span[@class="label"][contains(text(), "酒店规模")]/following-sibling::span/strong/text()')
    hotel_size = hotel_size[0].strip() if hotel_size else ""

    hotel_provided_1 = html.xpath('//div[@class="info-section"][2]//dl[@class="clearfix"][1]//li[not(contains(@class, "off"))]/text()')
    hotel_provided_1 = hotel_provided_1 if hotel_provided_1 else []

    hotel_provided_2 = html.xpath('//div[@class="info-section"][2]//dl[@class="clearfix"][2]//li//text()')
    hotel_provided_2 = hotel_provided_2 if hotel_provided_2 else []

    hotel_provided_3 = html.xpath('//div[@class="info-section"][2]//dl[@class="clearfix"][3]//li//text()')
    hotel_provided_3 = hotel_provided_3 if hotel_provided_3 else []
    
    hotel_provided = hotel_provided_1 + hotel_provided_2 + hotel_provided_3
    hotel_provided = ','.join(hotel_provided)

    hotel_not_provided = html.xpath('//div[@class="info-section"][2]//dl[@class="clearfix"][1]//li[@class = "off"]/text()')
    hotel_not_provided = ','.join(hotel_not_provided) if hotel_not_provided else ""

    hotel_summary = html.xpath('//div[@class="info-section"][4]//dd//text()')
    hotel_summary = hotel_summary[0].strip() if hotel_summary else ""

    hotel_list['open_time'] = hotel_open_time
    hotel_list['out_time'] = hotel_out_time
    hotel_list['constructed_year'] = hotel_constructed_year
    hotel_list['renovated_year'] = hotel_renovated_year
    hotel_list['size'] = hotel_size
    hotel_list['provided'] = hotel_provided
    hotel_list['not_provided'] = hotel_not_provided
    hotel_list['summary'] = hotel_summary

    return hotel_list
```

```python
def ua():
    """随机获取一个浏览器用户信息"""
    user_agents = [
        'Mozilla/5.0 (Windows; U; Windows NT 5.1; it; rv:1.8.1.11) Gecko/20071127 Firefox/2.0.0.11',
        'Opera/9.25 (Windows NT 5.1; U; en)',
        'Mozilla/4.0 (compatible; MSIE 6.0; Windows NT 5.1; SV1; .NET CLR 1.1.4322; .NET CLR 2.0.50727)',
        'Mozilla/5.0 (compatible; Konqueror/3.5; Linux) KHTML/3.5.5 (like Gecko) (Kubuntu)',
        'Mozilla/5.0 (X11; U; Linux i686; en-US; rv:1.8.0.12) Gecko/20070731 Ubuntu/dapper-security Firefox/1.5.0.12',
        'Lynx/2.8.5rel.1 libwww-FM/2.14 SSL-MM/1.4.1 GNUTLS/1.2.9',
        'Mozilla/5.0 (X11; Linux i686) AppleWebKit/535.7 (KHTML, like Gecko) Ubuntu/11.04 Chromium/16.0.912.77 Chrome/16.0.912.77 Safari/535.7',
        'Mozilla/5.0 (X11; Ubuntu; Linux i686; rv:10.0) Gecko/20100101 Firefox/10.0',
    ]
    agent = random.choice(user_agents)
    return {
        'Accept': '*/*',
        'Accept-Encoding': 'gzip, deflate, br',
        'Accept-Language': 'zh-CN,zh;q=0.9',
        'content-lenght' : '185',
        'Connection': 'keep-alive',
        'Referer': 'https://you.ctrip.com/',
        'User-Agent': agent,
        'sec-ch-ua-mobile': '?0',
        'sec-ch-ua-platform': "macOS"
    }
for canteen_list in tqdm(canteen_lists_tmp[1381:]):
    canteen_url = canteen_list['url']
    r = requests.get(url=canteen_url, headers=ua())
    html = etree.HTML(r.text)
    canteen_food = html.xpath('//div[@class="text_style"]//p//text()')
    canteen_food = canteen_food[0].strip() if canteen_food else ""

    canteen_classic = html.xpath('//ul[@class="s_sight_in_list s_sight_noline cf"]/li[2]//a/text()')
    canteen_classic = ''.join(canteen_classic) if canteen_classic else ""

    # 获取第三个<li>里的“暂无”
    canteen_tel = html.xpath('//ul[@class="s_sight_in_list s_sight_noline cf"]/li[3]/span[@class="s_sight_con"]/text()')
    canteen_tel = canteen_tel[0].strip() if canteen_tel else ""

    # 获取第五个<li>里的“营业中 08:00-23:00”
    canteen_time = html.xpath('//ul[@class="s_sight_in_list s_sight_noline cf"]/li[5]/span[@class="s_sight_con"]/text()')
    canteen_time = canteen_time[0].strip() if canteen_time else ""

    canteen_list['food'] = canteen_food.replace(" ", "").replace("\r\n", "")
    canteen_list['classic'] = canteen_classic
    canteen_list['tel'] = canteen_tel
    canteen_list['time'] = canteen_time
```
